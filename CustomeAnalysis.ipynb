```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.preprocessing import StandardScaler
```

# Load data

```python
data = pd.read_csv('marketing_campaign.csv', sep='\t')
```

```python
data.head()
```

```python
data.info()
data['Marital_Status'].value_counts()
```

```python
data['Age'] = 2014 - data['Year_Birth']
data['Total_Spent'] = data['MntWines'] + data['MntFruits'] + data['MntMeatProducts'] + data['MntFishProducts'] + data['MntSweetProducts'] + data['MntGoldProds']
data['Spending_Ratio'] = data['Total_Spent'] / data['Income']
data['Children_Home'] = data['Kidhome'] + data['Teenhome']
data['Marital_Status'] = data['Marital_Status'].replace(['Alone', 'Absurd', 'YOLO', 'Divorced', 'Widow', 'Single'], 0)
data['Marital_Status'] = data['Marital_Status'].replace(['Married', 'Together'], 1)
data = data.rename(columns={'Marital_Status': 'Has_Partner'})
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'], format='%d-%m-%Y', errors='coerce')
ref_date = data["Dt_Customer"].max()
data["Customer_Tenure_Days"] = (ref_date - data["Dt_Customer"]).dt.days
data = data.dropna(subset=['Income'])
data['Income'] = data['Income'].astype(np.int64)
data = data.drop(columns=['ID', 'Year_Birth', 'Dt_Customer', 'Z_CostContact', 'Z_Revenue', 'Kidhome', 'Teenhome'])
data=data.rename(columns={'MntWines': "Wines",'MntFruits':'Fruits','MntMeatProducts':'Meat','MntFishProducts':'Fish','MntSweetProducts':'Sweets','MntGoldProds':'Gold'})

data.head()

#data.to_csv("cleaned_marketing_campaign.csv", index=False)



```

**Purpose:** Reveal patterns that support  
- **Goal 1:** Response prediction (what drives *Response* = 1)  
- **Goal 3:** Personas/clusters (distinct customer groups)

**Input**
- `data/customers_featured.csv` (from Notebook 01)

**Outputs**
- At least 10 figures in `reports/figs/`  
- Short notes per figure (captions inside notebook)  
- Optional summary tables in `reports/tables/`


```python

# Imports & paths
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

DATA_IN = "customers_featured.csv"
FIG_DIR = Path("reports/figs")
TAB_DIR = Path("reports/tables")
FIG_DIR.mkdir(parents=True, exist_ok=True)
TAB_DIR.mkdir(parents=True, exist_ok=True)

pd.set_option("display.max_columns", 200)
sns.set_style("whitegrid")
```

```python
#Global styling
# Professional EDA Plotting Configuration

# ============================================================
# GLOBAL STYLE CONFIGURATION
# ============================================================

# Global styling configuration
rcParams.update({
    'figure.facecolor': 'white',
    'figure.dpi': 100,
    
    'axes.facecolor': 'white',
    'axes.edgecolor': '#E0E0E0',
    'axes.linewidth': 1.5,
    'axes.grid': True,
    'axes.grid.axis': 'y',
    'axes.axisbelow': True,
    'axes.labelsize': 11,
    'axes.labelweight': '600',
    'axes.labelcolor': '#2C3E50',
    'axes.titlesize': 14,
    'axes.titleweight': 'bold',
    'axes.titlepad': 20,
    'axes.titlecolor': '#1A1A1A',
    'axes.spines.top': False,
    'axes.spines.right': False,
    
    'grid.color': '#F0F0F0',
    'grid.linestyle': '-',
    'grid.linewidth': 1,
    'grid.alpha': 1,
    
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'xtick.color': '#4A4A4A',
    'ytick.color': '#4A4A4A',
    'xtick.major.size': 0,
    'ytick.major.size': 0,
    
    'legend.fontsize': 10,
    'legend.frameon': True,
    'legend.framealpha': 1,
    'legend.edgecolor': '#E0E0E0',
    'legend.facecolor': 'white',
    
    'font.family': 'sans-serif',
    'font.sans-serif': ['Arial', 'Helvetica', 'DejaVu Sans'],
    
    'savefig.dpi': 150,
    'savefig.bbox': 'tight',
    'savefig.facecolor': 'white',
})

# Modern, professional color palette
PALETTE = {
    'primary': '#5B7C99',      # Muted slate blue
    'secondary': '#8B7D6B',    # Warm taupe
    'accent1': '#C17767',      # Terracotta
    'accent2': '#7FA99B',      # Sage green
    'accent3': '#D4A76A',      # Golden sand
    'neutral': '#95A3B3',      # Cool gray
    
    'no_response': '#C17767',  # Terracotta (warm)
    'response': '#7FA99B',     # Sage (cool)
}

# Multi-color palettes for categorical data
CATEGORICAL_PALETTE = ['#5B7C99', '#8B7D6B', '#C17767', '#7FA99B', '#D4A76A', '#95A3B3']

sns.set_style("white")
sns.set_palette(CATEGORICAL_PALETTE)

# ============================================================
# UTILITY FUNCTIONS
# ============================================================

def annotate_bars(ax, fmt="{:.1f}", offset=3, fontsize=10, fontweight='bold'):
    """Add clean value labels on bars"""
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.text(p.get_x() + p.get_width()/2., height + offset,
                   fmt.format(height), 
                   ha='center', va='bottom', 
                   fontweight=fontweight, fontsize=fontsize, color='#2C3E50')

def format_currency(x, pos=None):
    """Format as currency"""
    if x >= 1000:
        return f'${x/1000:.0f}K'
    return f'${x:.0f}'

def finalize(ax, filename, dpi=150):
    """Apply final touches and save"""
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color('#E0E0E0')
    ax.spines['bottom'].set_color('#E0E0E0')
    
    plt.tight_layout()
    plt.savefig(FIG_DIR / filename, dpi=dpi, bbox_inches='tight', facecolor='white')
    plt.show()

def add_stats_box(ax, text, position='upper right'):
    """Add statistics box"""
    positions = {
        'upper right': (0.97, 0.97),
        'upper left': (0.03, 0.97),
        'lower right': (0.97, 0.03),
        'lower left': (0.03, 0.03)
    }
    x, y = positions.get(position, (0.97, 0.97))
    
    ax.text(x, y, text, transform=ax.transAxes,
            verticalalignment='top' if 'upper' in position else 'bottom',
            horizontalalignment='right' if 'right' in position else 'left',
            bbox=dict(boxstyle='round,pad=0.6', facecolor='white', 
                     edgecolor='#E0E0E0', linewidth=1.5),
            fontsize=9, family='monospace', color='#4A4A4A')

```

```python

# Load
df = pd.read_csv(DATA_IN)
print("Loaded:", DATA_IN, "â†’", df.shape)
display(df.head(3))
display(df.isnull().mean().sort_values(ascending=False).head(10))
```

```python
# Class balance & naive baseline
resp = df["Response"].map({0:"No", 1:"Yes"}).value_counts(normalize=True).mul(100)
fig, ax = plt.subplots(figsize=(9, 5.5))

bars = ax.bar(range(2), resp.reindex(["No", "Yes"]).values, 
              color=[PALETTE['no_response'], PALETTE['response']], 
              width=0.5, zorder=3)

ax.set_xticks(range(2))
ax.set_xticklabels(['No Response', 'Responded'], fontsize=11, fontweight='600')
ax.set_ylabel('Percentage of Customers (%)', fontweight='600')
ax.set_title('Campaign Response Distribution', pad=20)
ax.set_ylim(0, max(resp.values) * 1.15)

annotate_bars(ax, fmt="{:.1f}%", offset=1.5)
finalize(ax, "fig00_response_distribution.png")
print(f"â†’ Baseline accuracy: {resp['No']:.1f}%\n")

```

## Figure 1 â€” Missingness heatmap

```python
missing_pct = df.isnull().mean() * 100
missing_pct = missing_pct[missing_pct > 0].sort_values(ascending=True)

if len(missing_pct) > 0:
    fig, ax = plt.subplots(figsize=(10, max(5, len(missing_pct)*0.35)))
    
    bars = ax.barh(range(len(missing_pct)), missing_pct.values, 
                   color=PALETTE['accent3'], height=0.6, zorder=3)
    
    ax.set_yticks(range(len(missing_pct)))
    ax.set_yticklabels([col.replace('_', ' ') for col in missing_pct.index])
    ax.set_xlabel('Missing Data (%)', fontweight='600')
    ax.set_title('Missing Data by Feature', pad=20)
    ax.set_xlim(0, max(missing_pct.values) * 1.15)
    
    # Smart label placement: inside for long bars, outside for short bars
    threshold = max(missing_pct.values) * 0.15  # 15% of max value
    
    for i, v in enumerate(missing_pct.values):
        if v > threshold:
            # Long bars: label inside on the right
            ax.text(v - 0.05, i, f'{v:.1f}%', va='center', ha='right',
                    fontweight='bold', fontsize=10, color='white')
        else:
            # Short bars: label outside on the right
            ax.text(v + 0.05, i, f'{v:.1f}%', va='center', ha='left',
                    fontweight='bold', fontsize=10, color='#2C3E50')
    
    finalize(ax, "fig01_missing_data.png")
    print(f"â†’ {len(missing_pct)} features with missing data\n")
else:
    print("â†’ No missing data\n")

# Note: If any column >30% missing, discuss imputation/drop strategy in report.
```

## Figure 2 â€” Age distribution

```python

age_data = df["Age"].dropna()
age_data = age_data[(age_data >= 18) & (age_data <= 100)]

fig, ax = plt.subplots(figsize=(11, 5.5))
ax.hist(age_data, bins=30, color=PALETTE['primary'], 
        edgecolor='white', linewidth=2, zorder=3)

ax.set_xlim(18, 100)
ax.set_xlabel('Customer Age (years)', fontweight='600')
ax.set_ylabel('Number of Customers', fontweight='600')
ax.set_title('Customer Age Distribution', pad=20)

stats_text = f"Mean:   {age_data.mean():.1f} years\nMedian: {age_data.median():.0f} years\nStd:    {age_data.std():.1f} years"
add_stats_box(ax, stats_text)

finalize(ax, "fig02_customer_age.png")
print(f"â†’ Age range: {age_data.min():.0f}-{age_data.max():.0f} years\n")
# Note: Life-stage shape; supports Age_Bin.
```

## Figure 3 â€” Income (log) distribution

```python

if "log_Income" in df.columns:
    income_data = df["log_Income"].dropna()
    
    fig, ax = plt.subplots(figsize=(11, 5.5))
    ax.hist(income_data, bins=35, color=PALETTE['accent2'], 
            edgecolor='white', linewidth=2, zorder=3)
    
    ax.set_xlabel('Log-Transformed Annual Income', fontweight='600')
    ax.set_ylabel('Number of Customers', fontweight='600')
    ax.set_title('Customer Income Distribution (Log Scale)', pad=20)
    
    finalize(ax, "fig03_income_distribution.png")
    print("â†’ Income distribution visualized\n")


# Note: Heavy tails â†’ prefer log for modeling & clustering.
```

## Figure 4 â€” Total Spent by Education (boxplot)

```python

if "Education" in df.columns and "Total_Spent" in df.columns:
    edu_order = ['Basic', '2n Cycle', 'Graduation', 'Master', 'PhD']
    edu_order = [e for e in edu_order if e in df["Education"].unique()]
    
    fig, ax = plt.subplots(figsize=(11, 6))
    
    bp_data = [df[df['Education'] == edu]['Total_Spent'].dropna() for edu in edu_order]
    
    bp = ax.boxplot(bp_data, positions=range(len(edu_order)), widths=0.5, 
                    patch_artist=True, showfliers=False,
                    boxprops=dict(linewidth=1.5, edgecolor='white'),
                    whiskerprops=dict(linewidth=1.5, color='#6C757D'),
                    capprops=dict(linewidth=1.5, color='#6C757D'),
                    medianprops=dict(linewidth=2.5, color='#2C3E50'))
    
    for patch, color in zip(bp['boxes'], CATEGORICAL_PALETTE[:len(edu_order)]):
        patch.set_facecolor(color)
    
    ax.set_xticks(range(len(edu_order)))
    ax.set_xticklabels(edu_order, rotation=15, ha='right', fontweight='600')
    ax.set_xlabel('Education Level', fontweight='600')
    ax.set_ylabel('Total Spending ($)', fontweight='600')
    ax.set_title('Customer Spending by Education Level', pad=20)
    ax.yaxis.set_major_formatter(plt.FuncFormatter(format_currency))
    
    finalize(ax, "fig04_spending_education.png")
    print(f"â†’ {len(edu_order)} education levels analyzed\n")
```

## Figure 5 â€” Response rate by Age bucket

```python

if "Age_Bin" in df.columns and "Response" in df.columns:
    response_rates = df.groupby("Age_Bin")["Response"].mean() * 100
    
    fig, ax = plt.subplots(figsize=(11, 5.5))
    bars = ax.bar(range(len(response_rates)), response_rates.values, 
                  color=PALETTE['primary'], width=0.6, zorder=3)
    
    ax.set_xticks(range(len(response_rates)))
    ax.set_xticklabels(response_rates.index, fontweight='600')
    ax.set_xlabel('Age Segment', fontweight='600')
    ax.set_ylabel('Response Rate (%)', fontweight='600')
    ax.set_title('Campaign Response Rate by Age Segment', pad=20)
    ax.set_ylim(0, response_rates.max() * 1.2)
    
    annotate_bars(ax, fmt="{:.1f}%", offset=0.8)
    finalize(ax, "fig05_response_by_age.png")
    print(f"â†’ Best segment: {response_rates.idxmax()} ({response_rates.max():.1f}%)\n")

```

## Figure 6 â€” Income vs Total Spent (colored by Response)

```python

if all(c in df.columns for c in ["Income", "Total_Spent", "Response"]):
    fig, ax = plt.subplots(figsize=(11, 7))
    
    no_resp = df[df["Response"] == 0]
    yes_resp = df[df["Response"] == 1]
    
    ax.scatter(no_resp["Income"], no_resp["Total_Spent"], 
              alpha=0.5, s=25, c=PALETTE['no_response'], 
              label='No Response', edgecolors='none', zorder=2)
    ax.scatter(yes_resp["Income"], yes_resp["Total_Spent"], 
              alpha=0.8, s=45, c=PALETTE['response'], 
              label='Responded', edgecolors='white', linewidths=1, zorder=3)
    
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('Annual Income ($, log scale)', fontweight='600')
    ax.set_ylabel('Total Spending ($, log scale)', fontweight='600')
    ax.set_title('Income vs. Spending by Campaign Response', pad=20)
    ax.legend(loc='upper left', framealpha=1)
    ax.grid(True, alpha=0.4, which='both')
    
    finalize(ax, "fig06_income_spending_response.png")
    print(f"â†’ {len(yes_resp)} responders vs {len(no_resp)} non-responders\n")
```

## Figure 7 â€” Correlation heatmap (numeric)

```python

numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
corr_matrix = df[numeric_cols].corr()

fig, ax = plt.subplots(figsize=(14, 11))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)

sns.heatmap(corr_matrix, mask=mask, cmap='RdBu_r', center=0, 
            square=True, linewidths=1, linecolor='white',
            cbar_kws={"shrink": 0.75, "label": "Correlation"},
            annot=False, ax=ax, vmin=-1, vmax=1)

ax.set_title('Feature Correlation Matrix', pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

finalize(ax, "fig07_correlation_matrix.png")
print(f"â†’ Correlation matrix: {len(numeric_cols)} features\n")


# Note: watch multicollinearity; look for features correlated with Response & Monetary.
```

## Figure 7.1 - â€œtop signalsâ€ table (abs corr with Response)

## Figure 8 â€” Channel shares vs Response

```python
# FIGURE 8: CHANNEL PREFERENCES BY RESPONSE
channel_cols = ['Web_Share', 'Store_Share', 'Catalog_Share', 'Deals_Share']
available_channels = [c for c in channel_cols if c in df.columns]

if available_channels and "Response" in df.columns:
    n_channels = len(available_channels)
    fig, axes = plt.subplots(1, n_channels, figsize=(4*n_channels, 5))
    if n_channels == 1:
        axes = [axes]
    
    for ax, channel in zip(axes, available_channels):
        # Prepare data
        data_no = df[df["Response"] == 0][channel].dropna()
        data_yes = df[df["Response"] == 1][channel].dropna()
        
        # Create boxplot
        bp = ax.boxplot([data_no, data_yes], positions=[0, 1], widths=0.5,
                        patch_artist=True, showfliers=False,
                        boxprops=dict(linewidth=1.5, edgecolor='white'),
                        whiskerprops=dict(linewidth=1.5, color='#6C757D'),
                        capprops=dict(linewidth=1.5, color='#6C757D'),
                        medianprops=dict(linewidth=2.5, color='#2C3E50'))
        
        # Color the boxes - FIX: use correct PALETTE keys
        bp['boxes'][0].set_facecolor(PALETTE['no_response'])
        bp['boxes'][1].set_facecolor(PALETTE['response'])
        
        ax.set_xticks([0, 1])
        ax.set_xticklabels(['No', 'Yes'], fontweight='600')
        ax.set_xlabel('Campaign Response', fontweight='600')
        ax.set_ylabel('Usage Share', fontweight='600')
        ax.set_title(channel.replace('_Share', '').replace('_', ' '), pad=15)
        ax.grid(axis='y', alpha=1)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
    
    fig.suptitle('Channel Preferences by Campaign Response', 
                fontsize=14, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.savefig(FIG_DIR / "fig08_channel_preferences.png", dpi=150, bbox_inches='tight')
    plt.show()
    print(f"â†’ {n_channels} channels analyzed\n")
else:
    print("â†’ Channel columns not found, skipping Figure 8\n")
```

## Figure 9 â€” Category share profiles (top spenders)

```python

category_shares = [c for c in df.columns if c.endswith('_Share')]

if len(category_shares) >= 3:
    top_spenders = df.nlargest(200, 'Total_Spent')
    
    fig, ax = plt.subplots(figsize=(11, 5.5))
    
    bp = ax.boxplot([top_spenders[col].dropna() for col in category_shares],
                    labels=[c.replace('_Share', '') for c in category_shares],
                    patch_artist=True, showfliers=False, widths=0.6)
    
    colors = sns.color_palette('Set2', len(category_shares))
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)
        patch.set_alpha(0.8)
        patch.set_edgecolor('white')
        patch.set_linewidth(1.5)
    
    ax.set_ylabel('Share of Total Spending', fontweight='semibold')
    ax.set_title('Category Spending Patterns (Top 200 Customers)', pad=15)
    plt.xticks(rotation=25, ha='right')
    
    finalize(ax, "fig09_top_spender_categories.png")
    print(f"â†’ Analyzed spending patterns across {len(category_shares)} categories")
else:
    print("â†’ Insufficient category columns, skipping Figure 9")
```

## Figure 10 â€” RFM heatmap (mean Response by Recency_Bin Ã— Frequency quintile)

```python

if "Response" in df.columns:
    freq_q = pd.qcut(df["Frequency_pm"].fillna(0), q=5, labels=[1,2,3,4,5], duplicates="drop")
    pivot = pd.pivot_table(
        df.assign(Fq=freq_q),
        index="Recency_Bin", columns="Fq",
        values="Response", aggfunc="mean"
    )
    plt.figure(figsize=(6,4))
    sns.heatmap(pivot, annot=True, fmt=".2f")
    plt.title("Mean Response by Recency bin Ã— Frequency quintile")
    plt.tight_layout()
    plt.savefig(FIG_DIR / "fig10_rfm_resp_heatmap.png", dpi=150)
    plt.show()
    pivot.to_csv(TAB_DIR / "rfm_response_heatmap.csv")
```

## Figure 11 â€” Mutual Information (top features w.r.t. Response)

```python

try:
    from sklearn.preprocessing import StandardScaler
    from sklearn.feature_selection import mutual_info_classif
    y = df["Response"].fillna(0).astype(int)
    feat_cols = [
        "Recency","Frequency_pm","Monetary","RFM_Total","Total_Purchases",
        "Web_Share","Store_Share","Catalog_Share","Deals_Share",
        "Premium_Tilt","Category_Concentration","Past_Accepted",
        "Income","Age","Customer_Tenure_Days","Spending_Ratio"
    ]
    feat_cols = [c for c in feat_cols if c in df.columns]
    X = df[feat_cols].fillna(0).values
    Xs = StandardScaler().fit_transform(X)
    mi = mutual_info_classif(Xs, y, random_state=RANDOM_SEED)
    mi_s = pd.Series(mi, index=feat_cols).sort_values(ascending=False)
    plt.figure(figsize=(8,6))
    mi_s.head(15).plot(kind="barh")
    plt.gca().invert_yaxis()
    plt.title("Mutual Information with Response")
    plt.tight_layout()
    plt.savefig(FIG_DIR / "fig11_mutual_information.png", dpi=150)
    plt.show()
    mi_s.to_csv(TAB_DIR / "mi_scores.csv")
except Exception as e:
    print("MI computation skipped:", e)
```

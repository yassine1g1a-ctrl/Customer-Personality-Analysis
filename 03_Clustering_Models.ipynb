# Deep Learning Clustering on Customer Personality Dataset

In this notebook, we apply deep unsupervised learning methods to cluster customers based on their characteristics and behaviors.

We will use two approaches:
1. **Autoencoder + K-Means** â€” to learn a compact representation of customers and cluster them in latent space.
2. **Deep Embedded Clustering (DEC)** â€” an advanced end-to-end deep clustering model.

Before training, we ensure the dataset is properly encoded, cleaned, and scaled.

```python
# Basic tools
import pandas as pd
import numpy as np

# Preprocessing
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Deep learning
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Layer
from tensorflow.keras import backend as K

# Clustering
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE

import warnings
warnings.filterwarnings('ignore')
```

## 2. Encode categorical features and select relevant columns

We only keep meaningful numeric features for clustering.  
Derived columns (like log-transformed, binned, or interaction terms) are removed to avoid redundancy.


```python
# Drop columns that are derived or redundant
cols_to_drop = [
    'log_Income', 'log_Total_Spent', 'Income_x_HasPartner', 'Age_x_WebShare',
    'Age_Bin', 'Income_q', 'Recency_Bin', 'R_score', 'F_score', 'M_score', 'RFM_Total'
]

df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

# One-hot encode Education if it's not numeric
if 'Education' in df.columns and df['Education'].dtype == 'object':
    df = pd.get_dummies(df, columns=['Education'], drop_first=True)

df.shape, df.filter(like='Education').head()
```

## 3. Scale the features

We standardize all features (mean 0, variance 1) to ensure fair contribution to the model.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(X_scaled, columns=df.columns)

X_scaled.shape, df_scaled.head()
```

## 4. Method 1 : Autoencoder + K-Means

### 4.A- Build and train an Autoencoder

We train an autoencoder to compress the input features into a low-dimensional latent space.

- The encoder learns meaningful compressed representations.
- The decoder reconstructs the input to minimize reconstruction loss (MSE).

```python
input_dim = X_scaled.shape[1]
encoding_dim = 10  # latent space dimension (chosen after testing multiple values)

# Encoder
input_layer = Input(shape=(input_dim,))
encoder = Dense(64, activation='relu')(input_layer)
encoder = Dense(32, activation='relu')(encoder)
latent = Dense(encoding_dim, activation='relu', name='latent_space')(encoder)

# Decoder
decoder = Dense(32, activation='relu')(latent)
decoder = Dense(64, activation='relu')(decoder)
output_layer = Dense(input_dim, activation='linear')(decoder)

# Build model
autoencoder = Model(inputs=input_layer, outputs=output_layer)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

# Train
history = autoencoder.fit(X_scaled, X_scaled,
                          epochs=50,
                          batch_size=32,
                          validation_split=0.2,
                          verbose=1)
```

### 4.B- Extract the latent representations

We use the trained encoder to obtain the low-dimensional embeddings of each customer.

```python
encoder_model = Model(inputs=autoencoder.input, outputs=autoencoder.get_layer('latent_space').output)
X_latent = encoder_model.predict(X_scaled)
X_latent.shape
```

### 4.C- Apply K-Means on the latent space

We cluster the encoded representations and determine the optimal number of clusters using the Elbow Method.


```python
# Determine the optimal number of clusters
inertia = []
K_range = range(2, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_latent)
    inertia.append(kmeans.inertia_)

plt.plot(K_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
best_k = 5  # (set manually after checking the elbow curve)
kmeans = KMeans(n_clusters=best_k, random_state=42)
y_pred_kmeans = kmeans.fit_predict(X_latent)
```

### 4.D- Analyze and visualize clusters

```python
tsne = TSNE(n_components=2, random_state=42)
latent_2d = tsne.fit_transform(X_latent)

plt.figure(figsize=(8,6))
plt.scatter(latent_2d[:,0], latent_2d[:,1], c=y_pred_kmeans, cmap='tab10', s=6)
plt.title("Clusters - Autoencoder + KMeans")
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.show()
```

## 5. Method 2 : Deep Embedded Clustering (DEC)

In this section, we refine our clusters using **Deep Embedded Clustering (DEC)**, which optimizes the latent representation jointly with cluster assignments. This generally produces tighter, more meaningful clusters than standard K-Means.

The idea behind DEC is to jointly train:
- an **autoencoder** to learn compact latent representations, and  
- a **clustering layer** that refines cluster assignments during training.

### Concept

- Start with the encoder from Method 1 (Autoencoder)

- Initialize cluster centers using K-Means

- Refine clusters iteratively using a special Clustering Layer and KL divergence loss

- DEC helps the latent space directly reflect clustering structure.

### 5.A- Define the Clustering Layer

The DEC algorithm introduces a special *ClusteringLayer*, which computes a soft assignment of each data point to the cluster centroids using a Studentâ€™s t-distribution.

The idea is to make the model learn feature embeddings that are directly optimized for clustering.


```python
from tensorflow.keras.layers import Layer
from tensorflow.keras import backend as K

class ClusteringLayer(Layer):
    def __init__(self, n_clusters, weights=None, alpha=1.0, **kwargs):
        super().__init__(**kwargs)
        self.n_clusters = n_clusters
        self.alpha = alpha
        self.initial_weights = weights

    def build(self, input_shape):
        self.clusters = self.add_weight(
            shape=(self.n_clusters, input_shape[1]),
            initializer='glorot_uniform',
            name='clusters'
        )
        if self.initial_weights is not None:
            self.set_weights(self.initial_weights)
            self.initial_weights = None
        super().build(input_shape)

    def call(self, inputs):
        q = 1.0 / (1.0 + (K.sum(K.square(K.expand_dims(inputs, axis=1) - self.clusters), axis=2) / self.alpha))
        q **= (self.alpha + 1.0) / 2.0
        q = K.transpose(K.transpose(q) / K.sum(q, axis=1))
        return q
```

### 5.B- Build and Train the DEC Model

We now build the DEC model:
1. It takes the same latent representation from the encoder.
2. Uses the ClusteringLayer for soft assignments.
3. Iteratively refines the cluster centers by minimizing the KL divergence between the current assignment and a target distribution.

We start by initializing cluster centers with K-Means results from Method 1.


```python
from tensorflow.keras.models import clone_model

# Clone the encoder
encoder_model_DEC = clone_model(encoder_model)
encoder_model_DEC.set_weights(encoder_model.get_weights())
X_latent_dec = encoder_model_DEC.predict(X_scaled)

# Initialize clustering layer with K-Means centers
cluster_centers = kmeans.cluster_centers_ 
cluster_centers += 0.01 * np.random.randn(*kmeans.cluster_centers_.shape) # perturbation
clustering_layer = ClusteringLayer(best_k, weights=[cluster_centers], name='clustering')(encoder_model_DEC.output)
dec_model = Model(inputs=encoder_model_DEC.input, outputs=clustering_layer)
dec_model.compile(optimizer=Adam(0.001), loss='kld')

# Target distribution function
def target_distribution(q):
    weight = q ** 2 / q.sum(0)
    return (weight.T / weight.sum(1)).T

# DEC iterative training
maxiter = 10000
update_interval = 50
tol = 0.0001
y_pred_last = y_pred_kmeans

delta_history = []

for ite in range(maxiter):
    if ite % update_interval == 0:
        q = dec_model.predict(X_scaled, verbose=0)
        p = target_distribution(q)
        y_pred_dec = q.argmax(1)

        delta_label = np.mean(y_pred_dec != y_pred_last)
        delta_history.append(delta_label)
        y_pred_last = y_pred_dec

        print(f"Iter {ite}: Î”label = {delta_label:.5f}")
        
        if delta_label < tol:
            print("Converged early!")
            break

    dec_model.train_on_batch(X_scaled, p)

# After training
plt.plot(delta_history)
plt.title("Î”label evolution across iterations")
plt.xlabel("Update step")
plt.ylabel("Î”label")
plt.show()

```

### 5.C- Visualizing the Refined Clusters

After training, we visualize the refined latent space to observe improvements in cluster compactness and separation.


```python
# Refined latent space
tsne = TSNE(n_components=2, random_state=42)
latent_2d = tsne.fit_transform(X_latent_dec)

plt.figure(figsize=(8,6))
plt.scatter(latent_2d[:,0], latent_2d[:,1], c=y_pred_dec, cmap='viridis', s=6)
plt.title("DEC - Refined Latent Space Clusters")
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.show()
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Compute metrics
X_latent_dec = encoder_model_DEC.predict(X_scaled)

silhouette_final = silhouette_score(X_latent_dec, y_pred_dec)
calinski_final = calinski_harabasz_score(X_latent_dec, y_pred_dec)
davies_final = davies_bouldin_score(X_latent_dec, y_pred_dec)

print(f"DEC Silhouette Score: {silhouette_final:.4f}")
print(f"DEC Calinski-Harabasz Index: {calinski_final:.2f}")
print(f"DEC Davies-Bouldin Index: {davies_final:.4f}")

# Compare with Autoencoder + KMeans results
silhouette_kmeans = silhouette_score(X_latent, y_pred_kmeans)
calinski_kmeans = calinski_harabasz_score(X_latent, y_pred_kmeans)
davies_kmeans = davies_bouldin_score(X_latent, y_pred_kmeans)

print("\nAutoencoder + K-Means (baseline):")
print(f"Silhouette Score: {silhouette_kmeans:.4f}")
print(f"Calinski-Harabasz Index: {calinski_kmeans:.2f}")
print(f"Davies-Bouldin Index: {davies_kmeans:.4f}")

# --- Check how much assignments changed ---
changed_ratio = np.sum(y_pred_kmeans != y_pred_dec) / len(df)
print(f"\nFraction of points that changed cluster after DEC: {changed_ratio*100:.2f}%")
```
